## Loading data

```python
# Loading the raw datasets into corresponding pandas dataframes
filepath = r"WQD170053_Project_Churn_Prediction"
telco_demographics = pd.read_excel(filepath+r'\Telco_customer_churn_demographics.xlsx')
telco_services = pd.read_excel(filepath+r'\Telco_customer_churn_services.xlsx')
telco_status = pd.read_excel(filepath+r'\Telco_customer_churn_status.xlsx')
telco_location = pd.read_excel(filepath+r'\Telco_customer_churn_location.xlsx')
telco_population = pd.read_excel(filepath+r'\Telco_customer_churn_population.xlsx')
```

```python
# size of the tables in memory
print("Demographics:")
print(telco_demographics.info(verbose=False))
print("\nServices:")
print(telco_services.info(verbose=False))
print("\nStatus:")
print(telco_status.info(verbose=False))
print("\nLocation:")
print(telco_location.info(verbose=False))
print("\nPopulation:")
print(telco_population.info(verbose=False))
```

```python
#Combining the datasets
df1 = pd.merge(telco_demographics, telco_location.drop(columns="Count"), on ='Customer ID',how='left')
df2 = pd.merge(df1, telco_population.drop(columns="ID"), on ="Zip Code",how='left')
df3 = pd.merge(df2, telco_services.drop(columns="Count"), on ="Customer ID",how='left')
telco_df = pd.merge(df3, telco_status.drop(columns=["Count",'Quarter']), on ="Customer ID",how='left')
telco_df.head()
```
